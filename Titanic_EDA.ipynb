# Titanic Dataset Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load Dataset

```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

## Dataset Information

```python
df.info()
```

## Statistical Summary

```python
df.describe()
```

## Missing Values

```python
df.isnull().sum()
```

## Unique Values

```python
df.nunique()
```

## Value Counts

```python
df["Sex"].value_counts()

df["Pclass"].value_counts()

df["Survived"].value_counts()
```

## Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm"
            )
plt.show()
```

## Pairplot

```python
sns.pairplot(df,
             hue="Survived")

plt.show()
```

## Age Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"],
             bins=30,
             kde=True)

plt.show()
```

```python
## Fare Distribution
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Fare"])

plt.show()
```

## Age vs Fare

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Age",
    y="Fare",
    hue="Survived",
    data=df
)

plt.show()
```

```python
## Survival Count
```

```python
sns.countplot(
    x="Survived",
    data=df
)

plt.show()
```

## Survival by Gender

```python
sns.countplot(
    x="Sex",
    hue="Survived",
    data=df
)

plt.show()
```

## Passenger Class Distribution

```python
sns.countplot(
    x="Pclass",
    hue="Survived",
    data=df
)

plt.show()
```

## Embarked Port

```python
sns.countplot(
    x="Embarked",
    data=df
)

plt.show()
```

## Age by Passenger Class

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Pclass",
    y="Age",
    data=df
)

plt.show()
```

## Summary of Findings

1. Females had a much higher survival rate than males.

2. First-class passengers survived more frequently.

3. Fare has a positive relationship with survival.

4. Most passengers were between 20–40 years old.

5. Fare contains several outliers.

6. Cabin has many missing values.

7. Age has missing values.

8. Most passengers embarked from Southampton.
